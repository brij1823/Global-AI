```python
import pandas as pd
import numpy as np
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer

nltk.download('wordnet')
nltk.download('stopwords')

train_dataset = pd.read_csv("scotus_oral_arguments_train.csv")
val_dataset = pd.read_csv("scotus_oral_arguments_validation.csv")


def preprocess(df):
    tokenizer = RegexpTokenizer(r'\w+')
    stop = nltk.corpus.stopwords.words('english')
    lemm = WordNetLemmatizer()
    output = []
    for idx in df.index:
        print(f'Processing index: {idx}')
        timestamp = df['created_timestamp'][idx]
        text = df['text'][idx]
        texts = text.split('\n')
        for t in texts:
            s = t.split(':', 1)
            if len(s) == 2:
                speaker, lines = s[0], s[1]
                lines = [lemm.lemmatize(w) for w in tokenizer.tokenize(lines) if w not in stop]
                output.append([timestamp, speaker, lines])
    return output


train_preprocess = preprocess(train_dataset)
val_preprocess = preprocess(val_dataset)
```
